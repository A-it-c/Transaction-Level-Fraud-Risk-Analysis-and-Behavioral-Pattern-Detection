```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sample_data/creditcard.csv')
```

```python
df = df.dropna().reset_index(drop=True)
#print(df.isna().sum())
```

```python
df =df.drop_duplicates().reset_index(drop=True)
#print(f'Duplicated rows: {df.duplicated().sum()}')
```

```python
df['Class'] = df['Class'].astype('category')
df['Amount'] = df['Amount'].astype('float64')
df['Time'] = df['Time'].astype('int64')
#print(df.dtypes)
```

```python
df['Time_hours'] = df['Time'] / 3600
df['hour'] = ((df['Time'] // 3600) % 24).astype(int)
df['day'] = (df['Time'] // (3600 * 24)).astype(int)
df['time_segment'] = pd.cut(
df['hour'],
bins=[0, 6, 12, 18, 24],
labels=['Night', 'Morning', 'Afternoon', 'Evening'],
right=False
)
```

```python
df['log_Amount'] = np.log1p(df['Amount'])
amount_bins = [0, 10, 50, 100, 500, 1000, 5000, df['Amount'].max()]
df['amount_segment'] = pd.cut(
df['Amount'],
bins=amount_bins,
right=False
)
```

```python
df['is_fraud'] = df['Class'].astype(int)
df['fraud_amount'] = df['Amount'] * df['is_fraud']
```

```python
df['rule_high_amount'] = (df['Amount'] > 500).astype(int)
df['rule_night'] = (df['hour'] < 6).astype(int)
df['rule_combined'] = ((df['Amount'] > 500) & (df['hour'] < 6)).astype(int)
```

```python
df['true_positive'] = ((df['is_fraud'] == 1) & (df['rule_combined'] == 1)).astype(int)
df['false_positive'] = ((df['is_fraud'] == 0) & (df['rule_combined'] == 1)).astype(int)
df['false_negative'] = ((df['is_fraud'] == 1) & (df['rule_combined'] == 0)).astype(int)
```

```python
pca_features = [f'V{i}' for i in range(1, 29)]
```

```python
df.to_csv('creditcard_fraud.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10,5)
```

```python
#Fraud overviev. Доля шахрайства.
#Fraud нечаста подія (<0.2%), accuracy не підходить, фокус на recall/precision.

fraud_rate = df['is_fraud'].mean()

plt.bar(['Normal', 'Fraud'], df['is_fraud'].value_counts(normalize=True))
plt.title(f'Fraud Rate ~ {fraud_rate: .3%}')
plt.ylabel('Share')
plt.show()
```

```python
#Fraud по годинам. Шукаємо: нічні піки, підтвердження rule_night

hourly = df.groupby('hour')['is_fraud'].mean()

sns.lineplot(x=hourly.index, y=hourly.values)
plt.title('Fraud Rate by Hour')
plt.xlabel('Hour of day')
plt.ylabel('Fraud rate')
plt.show()
```

```python
#Transaction Amount Intelligence
# Fraud vs Non-Fraud (Amount)
#Висновок: Fraud характеризується або малими сумами, або екстремально великими.

sns.kdeplot(df[df.is_fraud==0]['log_Amount'], label='Normal', fill=True)
sns.kdeplot(df[df.is_fraud==1]['log_Amount'], label='Fraud', fill=True)

plt.title('Log Amount Distribution')
plt.legend()
plt.show()
```

```python
#Fraud за сегментами сум

amount_seg = df.groupby('amount_segment')['is_fraud'].mean().sort_values()

amount_seg.plot(kind='bar')
plt.title('Fraud Rate by Amount Segment')
plt.ylabel('Fraud rate')
plt.show()
```

```python
#Rule Engine Evaluation.  Precision rules. Якщо precision < 1% → noisy rule

rules = ['rule_high_amount', 'rule_night', 'rule_combined']

precision = {}
for r in rules:
  precision[r] = df[df[r]==1]['is_fraud'].mean()

  pd.Series(precision).plot(kind='bar')
  plt.title('Rule Precision')
  plt.ylabel('P(Fraud | Rule=1)')
  plt.show()
```

```python
#False Positives vs False Negatives

fp = df['false_positive'].sum()
fn = df['false_negative'].sum()

plt.bar(['False Positive', 'False Negative'], [fp, fn])
plt.title('Model Errors')
plt.show()
```

```python
#PCA-ознаки (V1–V28)

v_cols = [c for c in df.columns if c.startswith('V')]

fraud_means = df[df.is_fraud==1][v_cols].mean()
normal_means = df[df.is_fraud==0][v_cols].mean()

diff = (fraud_means - normal_means).abs().sort_values(ascending=False)

diff.head(10).plot(kind='bar')
plt.title('Top PCA Features (Fraud vs Normal)')
plt.show()
```
